Regression analysis is one of the most important fields in statistics and machine learning. There are many regression methods available. Linear regression is one of them.
 
 Simple Linear Regration-
 
linear regression is a linear approach to modelling the relationship between a scalar response and one or more explanatory variables Such models are called linear models.

```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
```

```python
data.shape
```

```python
data.describe()
```

```python
data.plot(x='Hours', y='Scores', style='o') 
plt.title('Hours vs Scores') 
plt.xlabel('Hours')  
plt.ylabel('Scores')  
plt.show()
```

from the graph above,we can clearly see that thare is a positive relation between number of hours and scores
and below graphcal representation of scores.


```python
plt.figure(figsize=(15,10))
plt.tight_layout()
seabornInstance.distplot(data['Scores'])
```

 preparing the data

the next step is to divide the data into attribute (inpute) and output(label)

```python
X = data['Hours'].values.reshape(-1,1)
y = data['Scores'].values.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

#training the data

We have split our data into training and testing sets.

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train) #training the algorithm
```

```python
#To retrieve the intercept:
print(regressor.intercept_)
#For retrieving the slope:
print(regressor.coef_)
```

```python
y_pred = regressor.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df
```

Making the Prediction

Now we train the algorithm to take some prediction. 

```python
df1 = df.head(25)
df1.plot(kind='bar',figsize=(16,10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
plt.scatter(X_test, y_test,  color='gray')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.show()
```

```python
#you can also predict the  own data
hours = 9.25
own_pred = regressor.predict([[hours]])
print("Number of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```

```python
from sklearn import metrics
print('Mean Absolute Error:',
     metrics.mean_absolute_error(y_test, y_pred))
```
